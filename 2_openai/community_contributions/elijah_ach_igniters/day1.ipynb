```python
import os
from dotenv import load_dotenv
from openai import AsyncOpenAI
from agents import Agent, Runner, trace, function_tool, OpenAIChatCompletionsModel, input_guardrail, GuardrailFunctionOutput
```

```python
load_dotenv(override=True)
```

```python
groq_client = AsyncOpenAI(base_url="https://api.groq.com/openai/v1", api_key=os.getenv("GROQ_API_KEY"))
llama = OpenAIChatCompletionsModel(model="llama-3.3-70b-versatile", openai_client=groq_client)
agent1 = Agent(name="Jokester", instructions="You are a joke teller", model=llama)
```

```python
with trace("Telling a joke"):
    result = await Runner.run(agent1, "Tell a joke about water.")
print(result.final_output)
```
