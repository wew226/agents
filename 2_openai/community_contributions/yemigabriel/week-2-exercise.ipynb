## Smart Meeting Scheduler Agent Built With The OpenAI Agents SDK

A user can describe a meeting in plain English, and the system will parse the request, check mock calendar availability, suggest the best time slots, confirm one slot, and optionally send calendar invites by email when the user clearly approves that.

**Agents**:

- A request parser agent turns natural language into structured meeting details
- A slot suggester agent uses tools to find common availability
- A coordinator/confirmer flow picks and confirms the best slot using a handoff
- An invite sender agent emails Calendar (.ics) invites through SendGrid
