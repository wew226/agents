# Ebenezer Casely Hayford — AI Personal Profiler

An interactive chatbot that answers questions about my career, background,
skills and experience.  
Business logic lives in **`profiler.py`** — this notebook only sets up the
Gradio interface.

```python
import gradio as gr

from profiler import Profiler
```

```python
profiler = Profiler(name="Ebenezer Casely Hayford")
```

```python
demo = gr.ChatInterface(
    fn=profiler.chat,
    type="messages",
    title="Chat with Eben",
    description=(
        "Ask me anything about my career, skills, background or experience. "
        "I'm happy to connect — feel free to share your email!"
    ),
    examples=[
        "What is your current role?",
        "Tell me about your technical skills.",
        "What projects are you most proud of?",
        "Are you open to new opportunities?",
    ],
    theme=gr.themes.Soft(),
)

demo.launch(inbrowser=True)
```
