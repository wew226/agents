```python
# First let's do an import. If you get an Import Error, double check that your Kernel is correct..

from dotenv import load_dotenv
from IPython.display import Markdown, display
import os
from openai import OpenAI

load_dotenv(override=True)

openai_api_key = os.getenv('OPENAI_API_KEY')
openai = OpenAI()
```

```python

# First create the messages:

messages = [{"role": "user", "content": "Pick a buisness area that might be worth exporing for an Agentic AI opportunity! Answer in one sentence!"}]

# Then make the first call:

response = openai.chat.completions.create(
    model = "gpt-4.1-mini",
    messages=messages
)

# Then read the business idea:

business_idea = response.choices[0].message.content
# print(business_idea)

# And repeat! In the next message, include the business idea within the message

message_pain_point = [{"role": "user", "content":f"Present the pain points fo the following business area what was suggested to that might be worth exporing for an Agentic AI opportunity: {business_idea}"}]

response_pain_point = openai.chat.completions.create(
    model="gpt-4.1-mini",
    messages=message_pain_point
)

pain_point = response_pain_point.choices[0].message.content
# print(pain_point)

message_propose_agentic_solution = message_pain_point = [{"role": "user", "content":f"""
                                                          Propose AI solution for the following business area what was suggested to that might be worth exporing for an Agentic AI opportunity: {business_idea}.
Consider the following pain points as well: {pain_point}
"""}]

response_propose_agentic_solution = openai.chat.completions.create(
    model="gpt-4.1-mini",
    messages=message_propose_agentic_solution
)

proposed_agentic_ai_solution = response_propose_agentic_solution.choices[0].message.content
display(Markdown(proposed_agentic_ai_solution))
```

```python
# Shorter version the gpt_answer_generator initiates a new agent every time in the loop

def gpt_answer_generator(message:list)->list:
    response = openai.chat.completions.create(
    model="gpt-4.1-mini",
    messages=message
).choices[0].message.content
    message.append({"role":"assistant", "content":response})
    return message

questions=[
    "Pick a buisness area that might be worth exporing for an Agentic AI opportunity!",
    "Present the pain points of this business area",
    "Propose AI solution for the following business area, consider the pain points as well."
]

message = []
for q in questions:
    message.append({"role": "user", "content":q})
    message = gpt_answer_generator(message)

display(Markdown(message[-1].get("content")))
```
