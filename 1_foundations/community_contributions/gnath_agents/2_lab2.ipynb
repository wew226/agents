## Agentic Workflow - Block Diagram

This notebook uses a multi-model agentic workflow to generate a technical presentation draft, collect specialized feedback, and produce a final consolidated summary.

```text
+---------------------------+
| 1) User Input / Prompt    |
| Topic + constraints       |
+-------------+-------------+
              |
              v
+---------------------------+
| 2) Draft Generator Agent  |
| LLM creates slide draft   |
+-------------+-------------+
              |
              v
+---------------------------+        +-----------------------------+
| 3) Draft Output           |------->| 4A) Technical Reviewer     |
| (presentation content)    |        | Accuracy, depth, rigor      |
+-------------+-------------+        +--------------+--------------+
              |                                       |
              |                                       v
              |                        +-----------------------------+
              |                        | Technical Feedback          |
              |                        +--------------+--------------+
              |
              |                        +-----------------------------+
              +----------------------->| 4B) Slide/Format Reviewer  |
                                       | Clarity, flow, visuals      |
                                       +--------------+--------------+
                                                      |
                                                      v
                                       +-----------------------------+
                                       | Format Feedback             |
                                       +--------------+--------------+
                                                      |
                                                      v
+---------------------------+          +-----------------------------+
| 5) Feedback Summarizer    |<---------| Both Feedback Streams       |
| Consolidates into actions |
+-------------+-------------+
              |
              v
+---------------------------+
| 6) Final Recommendations  |
| Prioritized improvement   |
+---------------------------+
```

Creating the Presentation Draft on AI Governance in English

````python
request = "Create a presentation draft on AI governance in English"
#request += "Answer only with the question, no explanation."
messages = [{"role": "user", "content": request}]
````

````python
messages
````

Using OpenAI gpt-4o-mini LLM model

````python
openai = OpenAI()
response = openai.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages,
)
ppt_draft = response.choices[0].message.content
print(ppt_draft)
````

Feedback 1: Evaluating the presentation draft content on technical perspective

````python
llm_models = []
feedback = []
feedback_on_tech_content = f"evaluate the presentation draft {ppt_draft} content on technical perspective and provide the feedback"
messages = [{"role": "user", "content": feedback_on_tech_content}]
````

Using OpenAI gpt-5-nano LLM model

````python


if "llm_models" not in globals():
    llm_models = []
if "feedback" not in globals():
    feedback = []
messages = [{"role": "user", "content": feedback_on_tech_content}]

model_name = "gpt-5-nano"

response = openai.chat.completions.create(model=model_name, messages=messages)
answer = response.choices[0].message.content

display(Markdown(answer))
llm_models.append(model_name)
feedback.append(answer)
````

Feedback 2: Feedback on Slide creation and format perspective

````python
feedback_on_slide_creation = f"evaluate the presentation draft {ppt_draft} content on slide creation and formatting perspective and provide the feedback"
messages = [{"role": "user", "content": feedback_on_slide_creation}]
````

Using Ollama model for generating Feedback-2

````python
ollama = OpenAI(base_url='http://localhost:11434/v1', api_key='ollama')
model_name = "llama3.2"

response = ollama.chat.completions.create(model=model_name, messages=messages)
answer = response.choices[0].message.content

display(Markdown(answer))
llm_models.append(model_name)
feedback.append(answer)
````

````python
# So where are we?

print(llm_models)
print(feedback)
````

````python
# It's nice to know how to use "zip"
for llm_model, feedback in zip(llm_models, feedback):
    print(f"Competitor: {llm_model}\n\n{feedback}")
````

````python
# Let's bring this together - note the use of "enumerate"

together = ""
for index, feedback in enumerate(feedback):
    together += f"# Response from competitor {index+1}\n\n"
    together += feedback + "\n\n"
````

````python
print(together)
````

Aggregating the feedbacks using LLM:

````python
aggregator = f"""You are aggregating the feedback from {len(llm_models)} competitors.
Each model has been given the feedback:

Now respond by aggregating the feedback from both the models in {together} and provide a consolidated feedback, nothing else. Do not include markdown formatting or code blocks."""
````

````python
print(aggregator)
````

````python
aggregator_messages = [{"role": "user", "content": aggregator}]
````

````python
# Judgement time!

openai = OpenAI()
response = openai.chat.completions.create(
    model="gpt-5.4-nano",
    messages=aggregator_messages,
)
feedback_summary = response.choices[0].message.content
print(feedback_summary)

````
