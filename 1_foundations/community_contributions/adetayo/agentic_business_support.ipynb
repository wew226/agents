This task introduces the agentic flow

The goal is to have an llm suggest business areas worth exploring for agentic oppurtunity, then pass the suggestion again to an llm to have it pickout a pain point that is ripe for agentic Ai solution and finally pass the response to an llm to propose the agentic AI solution

first the imports

```python
from dotenv import load_dotenv
import os
from openai import OpenAI
```

Load API Keys

```python
load_dotenv(override=True)
```

Lets setup openAI and the keys

```python
openai_api_key = os.getenv('OPENAI_API_KEY')
openai = OpenAI()
```

First prompt for business area

```python
businessAreareQuestion = "What are the business areas worth exploring for agentic oppurtunity. Reply with one business area only, no other text"
businessAreaMessage  = [{"role": "user", "content": businessAreareQuestion}]
businessAreaCompletion = openai.chat.completions.create(
    model="gpt-4.1-mini",
    messages=businessAreaMessage,
)
businessAreaResponse = businessAreaCompletion.choices[0].message.content
businessAreaResponse
```

Prompt for pain point in the industry

```python
businessAreaPainPointQuestion = f"""We have determined the the business area: {businessAreaResponse} worth exploring for agentic oppurtunit. Give me only one the pain point in this business area. Reply with only the pain point, no other text"""
businessAreaPainPointmessage  = [{"role": "user", "content": businessAreaPainPointQuestion}]
businessAreaPainPointResponseCompetion = openai.chat.completions.create(
    model="gpt-4.1-mini",
    messages=businessAreaPainPointmessage,
)
businessAreaPainPointResponse = businessAreaPainPointResponseCompetion.choices[0].message.content
businessAreaPainPointResponse
```

Prompt for Agebtic AI solution

```python
agenticAiSolutionQuestion = f"""We have determined the pain point: {businessAreaPainPointResponse} in the industry: {businessAreaResponse} is worth exploring for agentic oppurtunity. What is the agentic AI solution for this pain point."""
agenticAiSolutionMessage = [{"role": "user", "content": agenticAiSolutionQuestion}]
agenticAiSolutionResponse = openai.chat.completions.create(
    model="gpt-4.1-mini",
    messages=agenticAiSolutionMessage,
)

agenticAiSolutionResponse.choices[0].message.content
```
